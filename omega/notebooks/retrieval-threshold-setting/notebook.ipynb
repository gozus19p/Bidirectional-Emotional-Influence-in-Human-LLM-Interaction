```python
import pandas as pd

data: pd.DataFrame = pd.read_json("distances.json", orient="records")
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

distances = [entry['distance'] for entry in data.to_dict(orient="records")]

# Crea il grafico
plt.figure(figsize=(10, 6))
sns.histplot(distances, kde=True, color="blue", bins=15)

# Aggiungi etichette e titolo
plt.title('Distances Distribution')
plt.xlabel('Cosine Similarity')
plt.ylabel('Frequence')

# Mostra il grafico
plt.show()
```
